```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize'] = 8, 16
```

```python
#download the pretrained model save in to the variable reader
reader = easyocr.Reader(['en'])
```

```python

Image("/content/drive/MyDrive/EasyOCR/APJ.jpg")
```

```python
output = reader.readtext('/content/drive/MyDrive/EasyOCR/APJ.jpg')
output
```

```python
last_output = output[-3][0]
last_output
```

```python
x_min, y_min = [int(min(idx)) for idx in zip(*last_output)]
```

```python
x_max, y_max = [int(max(idx)) for idx in zip(*last_output)]
```

```python
image = cv2.imread('/content/drive/MyDrive/EasyOCR/APJ.jpg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
